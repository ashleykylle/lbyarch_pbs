```python
import pandas as pd
import struct
import math as math
df = pd.read_csv('FirstGenPokemon.csv')
display(df.head())
```

```python
type_conversion = {
    'None': 99,
    'normal': 0,
    'fighting': 1,
    'flying': 2,
    'poison': 3,
    'ground': 4,
    'rock': 5,
    'bug': 6,
    'ghost': 7,
    'steel': 8,
    'fire': 9,
    'water': 10,
    'grass': 11,
    'electric': 12,
    'psychic': 13,
    'ice': 14,
    'dragon': 15,
    'dark': 16,
    'fairy': 17
}
```

```python
# Function to convert types to their corresponding numbers
def convert_type(type_value):
    return type_conversion[type_value] if type_value in type_conversion else -1
```

```python
# Extract the relevant columns
df_extracted = df[["Number", " Type1", " Type2"]]
display(df_extracted.tail())
```

```python
# Replace NaN values with None
df_extracted[' Type1'] = df_extracted[' Type1'].fillna("None")
df_extracted[' Type2'] = df_extracted[' Type2'].fillna("None")
```

```python
# Replace the type strings with their corresponding numbers
df_extracted[' Type1'] = df_extracted[' Type1'].apply(convert_type)
df_extracted[' Type2'] = df_extracted[' Type2'].apply(convert_type)
display(df_extracted.head())
```

```python
# Create the desired format
df_extracted['Formatted'] = df_extracted.apply(lambda row: f"{row['Number']}{row[' Type1']}{row[' Type2']}", axis=1)
```

```python
# Output the result to separate text files
# Function to write a column to a binary file
def write_to_binary_file(filename, data):
    with open(filename, 'wb') as f:
        for value in data:
            f.write(struct.pack('I', value))  # 'i' format for a 4-byte integer

 # Write each column to a separate binary file
write_to_binary_file('pokemon_type1.bin', df_extracted[' Type1'])
write_to_binary_file('pokemon_type2.bin', df_extracted[' Type2'])
 
     
```

```python
# Function to read and print the contents of a binary file
def read_from_binary_file(filename):
    with open(filename, 'rb') as f:
        while chunk := f.read(4):  # Read 4 bytes at a time (size of an unsigned integer)
            value = struct.unpack('I', chunk)[0]
            print(value)

print("Numbers:")
read_from_binary_file('pokemon_type2.bin')

```
